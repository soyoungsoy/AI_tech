# subplot() vs subplots()

    - plt.subplot(row, col, idx) : fig 하나에 단수 ax
    - plt.subplots(nrows, ncols) : fig 하나에 복수 ax
    

```python
#subplot은 사이즈를 따로 빼서 쓴다.
plt.figure(figsize =(4,2))

plt.subplot(1,2,1)   #1줄 2칸 1번째
plt.bar([1,2,3],[10,20,30])

plt.subplot(1,2,2)   #1줄 2칸 2번째
plt.bar([1,2,3],[40,50,60])

plt.tight_layout
plt.show()
```

```python
                                        # subplot's'은 사이즈를 여기에 넣는다.
fig, axes = plt.subplots(nrows=1,ncols=2,figsize =(6,2))    #1행 2열
axes[0].bar([1,2,3],[10,20,30])

axes[1].bar([4,5,6],[30,20,40]) # subplots 는 axe가 여러개다.
plt.show()
```

## 4분위 이상치(Outlier) 탐지

```python
#이상치(outlier)는 무조건 지운다? = X
# 산 평균 온도에서 이상치 값이 나왔을 때 : drop 값이 아니라, 화재가 난 중요한 값이다.
# 데이터의 목적에 따라 값 처리를 결정한다. 
# 이상치 탐지를 목적으로 한 서비스가 있다.
```

<img src="https://img1.daumcdn.net/thumb/R1280x0/?scode=mtistory2&fname=https%3A%2F%2Fblog.kakaocdn.net%2Fdn%2FboDTe4%2FbtqHP3beJxg%2F8mzOK0wWK31QXpaY7VDJRk%2Fimg.png">

```python
df = pd.DataFrame(data=[1,4,    11,12,15,14,15,14,19,11,12,15,14,     77,70,80],
                  columns=['score']
                 )
print(df.head())

q1 = df["score"].quantile(q=0.25)  # 25%  해당 데이터 프레임의 quantile
q2 = df["score"].quantile(q=0.5)   # 중앙값
q3 = df["score"].quantile(q=0.75)  # 75%
IQR = q3 - q1
print (q1, q2, q3, IQR )  # 11.75   14.0    16.0    4.25

v_min = q1 - (IQR*1.5)
v_max = q3 + (IQR*1.5)
print (v_min,  v_max)     # 5.375   22.375
```

```python
df = pd.DataFrame(data=[1,4,    11,12,15,14,15,14,19,11,12,15,14,     77,70,80],
                  columns=['score']
                 )
print(df.head())


# q1 = df.quantile(q=0.25).values[0]  # 25%
q1 = np.percentile(df["score"], q=25)  # 25%
q2 = np.percentile(df["score"], q=50)  # 중앙값
q3 = np.percentile(df["score"], q=75)  # 75%
IQR = q3 - q1
print (q1, q2, q3, IQR )  # 11.75   14.0    16.0    4.25

v_min = q1 - (IQR*1.5)
v_max = q3 + (IQR*1.5)
print (v_min,  v_max)     # 5.375   22.375
```

```python
plt.boxplot(df['score'])
# plt.figure(figsize=(2,8))
plt.show()
```

* 이상치(Outlier)

```python
df[ (df['score']<v_min) | (df['score']>v_max) ]
```
