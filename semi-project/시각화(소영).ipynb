```python
fig = make_subplots(specs=[[{"secondary_y": True}]])

# Add traces
fig.add_trace(
    go.Scatter(x=df.index, y=df['한국은행_기준금리(월)'], name="한국은행_기준금리(월)"),
    secondary_y=True,
)

fig.add_trace(
    go.Bar(x=df.index, y=pre_bond.iloc[:,0:2], name='성별') ,
    secondary_y=False)


fig.add_trace(
    go.Bar(x=df.index,y = pre_bond.iloc[:,2:7], name='나이') ,
    secondary_y=False)


fig.update_yaxes(range=[70,100],secondary_y=False)
fig.show()
```
