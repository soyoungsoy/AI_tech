# Folium

```python
map = folium.Map(location=[37.5666805,126.9784147], zoom_start=14, tiles='OpenStreetMap')

tooltip = "클릭하면 장소명 보여줄게!"

folium.Marker(
    [37.5666805,126.9784147], popup="<i>시청</i>", tooltip=tooltip,
    icon=folium.Icon(color="red", icon="info-sign")
).add_to(map)

folium.CircleMarker(
    location=[37.5666805,126.9784147],
    radius=50,
    popup="시청 동네",
    color="#3186cc",
    fill=True,
    fill_color="#3186cc",
).add_to(map)

map
```

## Maker Cluster

* 마커들끼리 묶는 방법

```python
import folium
# from folium.plugins import MarkerCluster
```

```python
n = 50

lat_list = np.random.uniform(low = 36, high = 38, size = n) #랜덤 위도
lng_list = np.random.uniform(low = 127, high = 129, size = n)#램덤 경도

geo_list = []
for i, lat in enumerate(lat_list):
    geo_list.append((lat_list[i],lng_list[i]))
    

geo_list[:5]


name_list =[]

for i in range(n):
    name_list.append('빵집'+str(i))

map = folium.Map(location=[37.5666805,126.9784147], zoom_start=7, tiles='OpenStreetMap')    
    
plugins.MarkerCluster(geo_list, popups = name_list).add_to(map)
map
```
