```python

html_doc = """<html><head><title>The Dormouse's story</title></head>
<body>
<p class="title"><b>The Dormouse's story</b></p>

<p class="story">Once upon a time there were three little sisters; and their names were
<a href="http://example.com/elsie" class="sister" id="link1">Elsie</a>,
<a href="http://example.com/lacie" class="sister" id="link2">Lacie</a> and
<a href="http://example.com/tillie" class="sister" id="link3">Tillie</a>;
and they lived at the bottom of a well.</p>

<p class="story">...</p>
"""

soup = BeautifulSoup(html_doc, 'html.parser') #객체로 변환

print(soup.prettify())

#soup.태그[속성명] ex.soup.p['title']
```

```python
for link in soup.find_all('a'):
    print(link.get('href')) #a로 시작하는 곳의 href 가져와
    print(link.text) #a로 시작하는 것의 text 가져와
# http://example.com/elsie
# http://example.com/lacie
# http://example.com/tillie
```

* TEXT로 오는 데이터는 json이 필요없다. [[]], dict등의 방식으로 넣으면 된다.

```python
news_list=[] #데이터 프레임화 하려면 리스트의 리스트화해야 하기 때문에 리스트 정의하고, for문 append로 리스트에 담는다.


for aali in aaul :
    title = aali.select_one('a > div > span').text #제목 뽑아옴 셀렉트와 겹치는 부분 지우고 남긴다.
    regdate = aali.select_one('a > div > div > span.date').text
    contents = aali.select_one('a > div > div > span.desc').text
    href = aali.select_one('a').get('href')
    
    news_list.append([title,regdate,contents,href])
    
print([title,regdate,contents,href])

df = pd.DataFrame(news_list, columns=['title','regdate','contents','herf'])
df
```
