#left_on, Right_on : 키 값이 다를 때(e.empno = d.empno가 아닌 e.empno = d.emo 이런식일 때)

#컬럼으로 연결하든, 온으로 연결하든, 인덱스로 연결하든 하나만 선택해야한다. 다 설정하는 것이 아님.  
#sort 이하 기능은 보통 잘 안씀.

```python
left = pd.DataFrame(
    {
        "key": ["K0", "K1", "K2", "K3"],
        "A": ["A0", "A1", "A2", "A3"],
        "B": ["B0", "B1", "B2", "B3"],
    }
)

left
```

```python

right = pd.DataFrame(
    {
        "key": ["K0", "K1", "K2", "K3"],
        "C": ["C0", "C1", "C2", "C3"],
        "D": ["D0", "D1", "D2", "D3"],
    }
)

right
```

```python
result = pd.merge(left, right, on="key")

result
```

컬럼이름이 같으면 연결하는게 아니라, 값을 연결하는 것이다.  
우연히 키 이름이 똑같으면, on 해라.

```python
left = pd.DataFrame(
    {
        "key1": ["K0", "K0", "K1", "K2"],
        "key2": ["K0", "K1", "K0", "K1"],
        "A": ["A0", "A1", "A2", "A3"],
        "B": ["B0", "B1", "B2", "B3"],
    }
)


right = pd.DataFrame(
    {
        "key1": ["K0", "K1", "K1", "K2"],
        "key2": ["K0", "K0", "K0", "K0"],
        "C": ["C0", "C1", "C2", "C3"],
        "D": ["D0", "D1", "D2", "D3"],
    }
)


result = pd.merge(left, right, on=["key1", "key2"])

result
```

# inner join

<pre>
* 오라클 문법
select e.empno, e.ename, d.deptno, d.dname from emp e, dept d
where e.deptno= d.deptno;


* ANSI 문법
select e.empno, e.ename, d.deptno, d.dname from emp e JOIN dept d on e.deptno = d.deptno;

</pre>

```python
df = pd.merge(empdf, deptdf, on = "deptno", how = 'inner')
df[['empno','ename','deptno','dname']]
```

# Outer join - dept 기준

<pre>
* 오라클 문법
* Dept 기준 : 부족한 쪽에 (+)

select e.empno, e.ename, d.deptno, d.dname from emp e, dept d
where e.deptno<b>(+)</b>= d.deptno;


* ANSI 문법
* Dept 기준 : 왼쪽/ 오른쪽 누가 기준인지 정하기

select e.empno, e.ename, d.deptno, d.dname from emp e <b>RIGHT OUTER JOIN</b> dept d on e.deptno = d.deptno;

</pre>

```python
df = pd.merge(empdf, deptdf, on = "deptno", how = 'right')
df[['empno','ename','deptno','dname']]
```

## Full Outer Join

<pre>
