# L1 norm(Lasso)

    -벡터의 길이를 구하는 것은 놈을 통해 할 수 있습니다.
    - 아웃라이어 심한 경우. 쓸데없는 피쳐를 제거해버린다.
    - 알파를 이용해 0으로 만들어서 제거한다. 

# L2 norm(Ridge)
    
        - outlier가 큰 데이터에는 쓰면 안된다. 제곱하기 때문에.. 변형하기 때문에
        - 아웃라이어 민감하게 반응하기 때문에 사용X
        - 데이터의 값이 일반적인(아웃라이어가 별로 없는) 곳에 쓴다.
        - 0으로 만들지는 않는다.

# 규제 ★

    - 잔차제곱( = 회귀계수 W) 크기를 제어해서 과적합을 개선해 비용함수 목표를 알파로 조절하는 것
    - W가 크면 알파(규제값)을 낮춘다. (피쳐 영향력 축소)
    - W가 작으면 알파를 키워야. (피쳐 영향력 확대) 
    - 반비례 관계
